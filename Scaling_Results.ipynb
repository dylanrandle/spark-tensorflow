```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# "full dataset": video-level features
# first 1000 records for training
# first 100 records for testing

nodes=[1, 2, 4, 8, 15]                  # 16, 32, 64, 128 .... requested instance limit increase, but waiting
times_batch_async=[89, 36, 33, 30, 27]                # minutes
times_epoch_async=[82, 30, 13, 9.3, 3.5]               # minutes
accuracy=[0.65, 0.64, 0.64, 0.63]   # acc. slightly drops off
```

```python
speedup_batch=[times_batch_async[0]/t for t in times_batch_async]
speedup_epoch=[times_epoch_async[0]/t for t in times_epoch_async]
```

```python
# theoretical =  lambda p: (0.05 + (0.95/p) - 0.02)**-1
plt.figure(figsize=(10,6))
plt.plot(nodes,speedup_batch, '-o',label='batch-async')
plt.plot(nodes,speedup_epoch, '-o',label='epoch-async')
plt.plot(nodes,nodes, '-o', label='ideal')
plt.xlabel('Number of Worker Nodes', fontsize=15)
plt.ylabel('Speedup', fontsize=15)
plt.title('Experimental Speedup of YouTube-8M Classification', fontsize=15)
plt.legend()
plt.grid();
```
